As we can see this does not work, as we still have 1 null value in field named 'region'

This is because the following command does not make changes to the original dataframe (named 'df' in this case)

df['region'].fillna("NaN")



Inorder to make changes to the current dataframe(i.e dataframe named 'df' in this case) use the following command

df['region'].fillna("NaN", inplace= True) 


Another approach could pe to create a new dataframe named 'df2' and make changes there, rthis can be done via the following command:
